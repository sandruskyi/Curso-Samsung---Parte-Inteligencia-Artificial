### Ejercicio 1

Representad en una única gráfica las funciones f1 y f2 definidas más abajo, evaluadas en el intervalo [0, 2] y con paso (resolución) 0.01. Haced que los rangos de valores mostrados sean `[0.25, 2.25]` para el eje de las X y `[-1, 1]` para el eje de las Y. Adicionalmente, mostrad la gráfica correspondiente a ```f1``` con color <span style="color: green">verde</span> y la gráfica correspondiente a ```f2``` con color <span style="color: red">rojo</span>.



````python
import numpy as np

def f1(x):
    return np.exp(-x)

def f2(x):
    return np.exp(-x) * np.sin(2 * np.pi * x)
````

````python
# Respuesta
import matplotlib.pyplot as plt

# ¿Diferencia entre x=np.linspace(0,2,0.01) y x= np.arange(0,2,0.01)? 
x = np.arange(0,2, 0.01)

plt.plot(x, f1(x), 'g', x, f2(x), 'r') # Creamos el plot

# Rango de valores de x e y: 
plt.xlim(0.25, 2.25)
plt.ylim(-1,1)

# Etiquetas:
plt.xlabel('x')
plt.ylabel('y')

#Mostrar el plot:
plt.show()
````

````python
# Respuesta
# 2 gráficas de barras: los resultados del número de votos y del número de diputados para cada partido respectivamente.
# Una al lado de otra (horizontal), datos de manera horizontal (barras horizontales), EJE Y: etiquetas con el nomre de partidos, 
# a la derecha de cada barra: texto con el nº de votos para cada partido DENTRO del gráfico

import matplotlib.pyplot as plt
import numpy as np

partido = ['PSOE','PP','Cs','PODEMOS-IU-EQUO','VOX','ERC','ECP-GUANYEM','JxCAT','EAJ-PNV','PACMA']
votos = [7513142,4373653,4155665,2897419, 2688092	,1020392,615665,	500787,395884	,328299]
diputados = [123,66,57,33,24,15,7,7,6,0]

plt.subplot(1,3,1)
plt.barh(np.arange(len(votos)), votos, height = 0.8 ,align='center', tick_label = partido )
for i, v in enumerate(votos):
    plt.text(v + 3, i , str(v), color='green', fontweight='bold')

plt.subplot(1,3,3)
plt.barh(np.arange(len(diputados)), diputados, height=0.8, align='center', tick_label = partido)
for i, v in enumerate(diputados):
  plt.text(v+3,i, str(v), color='green',fontweight='bold')
````
